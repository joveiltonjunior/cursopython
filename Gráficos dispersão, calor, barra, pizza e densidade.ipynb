```python
# importat as bibliotecas que seram utilizada e atribui outro nome a elas para ficar mais limpo o codigo
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv('C:\\Users\\jovei\\OneDrive\\Área de Trabalho\\git\\EBAC\\Atividades\\ecommerce_preparados.csv')# comando para importat o arquivo 

df.head(10) #visualizar se foi importado tudo certo
```

```python
#Gráfico de dispersão 

plt.hexbin(df['Nota'],df['N_Avaliações'], gridsize= 30, cmap= 'Blues')# definir o tipo de gráfico a ser executado, eixo X e Y, o tamanho da grade e cor 
plt.colorbar(label = ' ')# 
plt.ylim([1, 1000])# Definir os limites do eixo X
plt.xlabel('Nota')# título do eixo x
plt.ylabel('Avaliações')# título do eixo y
plt.title("Dispersão Avaliações por Nota ")# título do gráfico 
plt.show #comando para mostrar o gráfico 
```

```python
# Mapa de Calor
mapa_de_calor = df[['Nota','N_Avaliações', 'Nota_MinMax', 'N_Avaliações_MinMax', 'Qtd_Vendidos_Cod', 'Marca_Freq' ]]# selecionas as variaveis para o grafico elas precisam ser numericas 
plt.figure(figsize=(10, 8))# dimeções do mapa de calor 
sns.heatmap(mapa_de_calor.corr(), annot=True, fmt=".2f", cmap="coolwarm")# Gerar o mapa de calor
plt.title("Mapa de Calor") # colocar o titulo do gráfico 
plt.show
```

```python
# Gráfico de Barras 


plt.figure(figsize=(10,6))
df['N_Avaliações'], df['Nota'].value_counts().plot(kind='bar', color='#90ee70')# Calcular as contagens e plotar
# Adicionar título e rótulo
plt.title('N_Avaliações -1')
plt.xlabel('N_Avaliações')
plt.ylabel('Nota')
plt.show


```

```python
# Gráfico de pizza 

# Criar uma nova coluna para agrupar notas menores que 3
df['Nota_Ajustada'] = df['Nota'].apply(lambda x: 'Menor que 4' if x < 4 else x)# Criar uma nova coluna para agrupar notas menores que 3
comb_counts = df.groupby(['Nota_Ajustada']).size()# Contar a frequência de cada categoria ('Menor que 4' e as demais notas)
# Obter os rótulos e valores
x = comb_counts.index  # Rótulos
y = comb_counts.values  # Valores
# Configurar o gráfico de pizza
plt.figure(figsize=(10, 6))  # Dimensões do gráfico de pizza
plt.pie(y, labels=x, autopct='%1.1f%%', startangle=90)
plt.title('Gráfico de Pizza - Notas (Agrupando Menor que 3)')# Adicionar título
plt.show()
```

```python
#Gráfico Densidade

plt.figure(figsize=(10,6)) # Dimensões do gráfico Densidade
sns.kdeplot(df['N_Avaliações'], fill= True, color= '#863e9c')# Gráfico de densidade para a coluna 'N_Avaliações'
# Configurações do gráfico
plt.title('Gráfico Densidade Avaliações')
plt.xlabel('Avaliações')
plt.show
```
